```python
import pandas as pd
print(pd.__version__)



dataframe = pd.read_csv('data.csv')
print(dataframe)

dataframe_customindex = dataframe.set_index('id')
print(dataframe_customindex)

print(dataframe)
dataframe_sorted = dataframe.sort_index()
print(dataframe_sorted)

dataframe_sorted = dataframe.sort_values(by = ['id','age'])
print(dataframe_sorted)

dataframe = dataframe_customindex.reset_index()
print(dataframe)

print(dataframe)

print()

O = dataframe['name']
print(O)

O = dataframe.iloc[:,1]
print(O)

O = dataframe.loc[:,'name']
print(O)

print(dataframe)

O = dataframe.iloc[[0,1],:]
print(O)

O = dataframe.loc[[0,1],:]
print(O)

print(dataframe)

dataframe_condition = dataframe.loc[dataframe.age >= 24]
print(dataframe_condition)

print(dataframe)
print()

occ_average_age = dataframe.groupby('occ')['age'].mean()
print(occ_average_age)

print(dataframe)
print()
dataframe_nullfill = dataframe.fillna(0)
print(dataframe_nullfill)
```
